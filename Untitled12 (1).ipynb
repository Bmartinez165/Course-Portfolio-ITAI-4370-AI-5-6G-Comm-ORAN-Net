```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

np.random.seed(42)
num_bits   = 8
bit_rate   = 1          
samples_per_bit = 100
fc         = 5          
SNR_dB     = 10

bits = np.random.randint(0, 2, num_bits)
print("Transmitted bits:", bits)

t_total = num_bits / bit_rate
t = np.linspace(0, t_total, num_bits * samples_per_bit, endpoint=False)

bpsk_symbols = 2 * bits - 1
symbol_signal = np.repeat(bpsk_symbols, samples_per_bit)
carrier       = np.cos(2 * np.pi * fc * t)
modulated     = symbol_signal * carrier

signal_power = np.mean(modulated**2)
SNR_linear   = 10 ** (SNR_dB / 10)
noise_power  = signal_power / SNR_linear
noise        = np.sqrt(noise_power) * np.random.randn(len(modulated))
received     = modulated + noise

demod = received * carrier          
demod_bits = []
for i in range(num_bits):
    segment = demod[i*samples_per_bit : (i+1)*samples_per_bit]
    decision = 1 if np.sum(segment) > 0 else 0
    demod_bits.append(decision)

demod_bits = np.array(demod_bits)
print("Received  bits:", demod_bits)
ber = np.sum(bits != demod_bits) / num_bits
print(f"Bit Error Rate: {ber:.2f}")

fig, axes = plt.subplots(4, 1, figsize=(12, 10))
fig.suptitle("BPSK Communication System – Signal Flow", fontsize=14, fontweight="bold", y=0.98)

t_bits = np.linspace(0, t_total, num_bits * samples_per_bit)
bit_wave = np.repeat(bpsk_symbols, samples_per_bit)
axes[0].step(t, bit_wave, where="post", color="#2196F3", linewidth=2)
axes[0].set_title("Source Signal (BPSK mapped: 0→-1, 1→+1)", fontsize=11)
axes[0].set_ylabel("Amplitude")
axes[0].set_ylim(-1.5, 1.5)
axes[0].set_xlim(0, t_total)
axes[0].axhline(0, color="gray", linewidth=0.5, linestyle="--")
axes[0].grid(True, alpha=0.3)

axes[1].plot(t, modulated, color="#4CAF50", linewidth=0.8)
axes[1].set_title(f"Modulated Signal (BPSK, carrier = {fc} Hz)", fontsize=11)
axes[1].set_ylabel("Amplitude")
axes[1].set_xlim(0, t_total)
axes[1].grid(True, alpha=0.3)

axes[2].plot(t, received, color="#FF5722", linewidth=0.6, alpha=0.85)
axes[2].set_title(f"Received Signal After AWGN Channel (SNR = {SNR_dB} dB)", fontsize=11)
axes[2].set_ylabel("Amplitude")
axes[2].set_xlim(0, t_total)
axes[2].grid(True, alpha=0.3)

recovered_wave = np.repeat(2 * demod_bits - 1, samples_per_bit)
original_wave  = np.repeat(bpsk_symbols, samples_per_bit)

axes[3].step(t, original_wave, where="post", color="#2196F3", linewidth=2.5, label="Original", alpha=0.6)
axes[3].step(t, recovered_wave, where="post", color="#9C27B0", linewidth=1.5,
             linestyle="--", label="Recovered")
axes[3].set_title(f"Original vs Recovered Bits  (BER = {ber:.2f})", fontsize=11)
axes[3].set_ylabel("Amplitude")
axes[3].set_xlabel("Time (s)")
axes[3].set_ylim(-1.5, 1.5)
axes[3].set_xlim(0, t_total)
axes[3].axhline(0, color="gray", linewidth=0.5, linestyle="--")
axes[3].legend(loc="upper right")
axes[3].grid(True, alpha=0.3)

for i, b in enumerate(bits):
    axes[0].text(i + 0.5, 1.25, str(b), ha="center", va="center",
                 fontsize=9, color="#1565C0", fontweight="bold")

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig("/mnt/user-data/outputs/signal_flow.png", dpi=150, bbox_inches="tight")
plt.show()
print("Plot saved to signal_flow.png")
```
